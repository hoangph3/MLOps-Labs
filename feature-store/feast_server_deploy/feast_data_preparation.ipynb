```python
import numpy as np
import pandas as pd
import pickle

with np.load("feature_repo/data/mnist.npz") as data:
#     train_examples = data['x_train']
#     train_labels = data['y_train']
    test_examples = data['x_test'].astype(np.float32)
    test_labels = data['y_test'].astype(np.int32)

print(test_examples.shape, test_labels.shape)
```

```python
# dump array to bytes
data = [{'feature': pickle.dumps(test_examples[i]), 'label_id': test_labels[i]} for i in range(len(test_examples))]
data_df = pd.DataFrame(data)
data_df
```

```python
# Creating timestamps for the data
timestamps = pd.date_range(
    end=pd.Timestamp.now(), 
    periods=len(data_df), 
    freq='H').to_frame(name="event_timestamp", index=False)

data_df = pd.concat(objs=[timestamps, data_df], axis=1)
data_df
```

```python
# Writing our DataFrames to parquet files
data_df.to_parquet(path='feature_repo/data/mnist.parquet')
```
