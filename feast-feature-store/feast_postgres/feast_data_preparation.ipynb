```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import json

with np.load("feature_repo/data/mnist.npz") as data:
#     train_examples = data['x_train'].astype(np.float64)
#     train_labels = data['y_train'].astype(np.int64)
    test_examples = data['x_test'].astype(np.float64)
    test_labels = data['y_test'].astype(np.int64)

print(test_examples.shape, test_labels.shape)
```

```python
data = []
for i in range(len(test_examples)):
    if i == 100:
        break
    example = json.dumps(test_examples[i].tolist())
    d = {}
    d['feature'] = example
#     for j in range(len(example)):
#         d[f'feature_{j}'] = example[j]
    d['label'] = test_labels[i]
    d['event_id'] = i
    data.append(d)
```

```python
data_df = pd.DataFrame(data)
data_df
```

```python
# Creating timestamps for the data
timestamps = pd.date_range(
    end=pd.Timestamp.now(), 
    periods=len(data_df), 
    freq='H').to_frame(name="event_timestamp", index=False)

data_df = pd.concat(objs=[timestamps, data_df], axis=1)
data_df
```
